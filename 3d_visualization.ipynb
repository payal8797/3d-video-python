```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
from PIL import Image
```

```python
#create two arrays
x = np.linspace(-20, 20, 200)
y = np.linspace(-15, 15, 200)
```

```python
#create two 2D arrays
X, Y = np.meshgrid(x, y)
```

```python
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(0.1*X) + np.cos(np.sqrt(X**2 + Y**2)) * np.sin(0.1*Y)
# Create the Figure and 3D Axes
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')
ax._axis3don = False

#3D Contour Plot
ax.contour3D(X, Y, Z, 100, cmap='viridis')
ax.view_init(60, 90)
```

```python
fileName = '3d'
# Create the '3d' directory if it doesn't exist
if not os.path.exists(fileName):
    os.makedirs(fileName)

#Generating and saving figures
for angle in range(0, 360):
    plt.figure(figsize=(16, 9))
    ax = plt.axes(projection='3d')
    ax._axis3don = False
    ax.contour3D(X, Y, Z, 200, cmap='viridis')
    ax.view_init(60, angle)
    
    # Save figure with different names depending on the view
    filename = fileName + '/' + 'vis_' + str(angle) + '.png'
    plt.savefig(filename, dpi=75)
    plt.close()
```

```python
png_count = 360
files = []
for i in range(png_count):
    seq = str(i)
    file_names = fileName + '/' + 'vis_' + seq +'.png'
    files.append(file_names)
    
print(files)
```

```python
# Create the frames
frames = []
for i in files:
    new_frame = Image.open(i)
    frames.append(new_frame)
    
# Save into a GIF file   
frames[0].save('3d_vis.gif', format='GIF',
               append_images=frames[1:],
               save_all=True,
               duration=40, loop=0)
```
