```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
from PIL import Image
```

```python
#create two arrays
x = np.linspace(-20, 20, 200)
y = np.linspace(-15, 15, 200)
```

```python
#create two 2D arrays
X, Y = np.meshgrid(x, y)
```

```python
Z = np.sin(np.sqrt(X**2 + Y**2)) * np.cos(0.1*X) + np.cos(np.sqrt(X**2 + Y**2)) * np.sin(0.1*Y)
# Create the Figure and 3D Axes
fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection='3d')
ax._axis3don = False

#3D Contour Plot
ax.contour3D(X, Y, Z, 300, cmap='viridis')
ax.view_init(60, 90)
```

```python
fileName = 'images'
# Create the 'images' directory if it doesn't exist
if not os.path.exists(fileName):
    os.makedirs(fileName)

#Generating and saving figures
for angle in range(0, 360):
    fig = plt.figure(figsize=(16, 9))
    ax = plt.axes(projection='3d')
    ax._axis3don = False
    ax.contour3D(X, Y, Z, 300, cmap='viridis')
    ax.view_init(60, angle)

    # Save figure with different names depending on the view
    filename = f'{fileName}/vis_{angle}.png'
    plt.savefig(filename, dpi=75, bbox_inches='tight', pad_inches=0)
    plt.close()
```

```python
images_count = 360
files = [f"{fileName}/vis_{i}.png" for i in range(images_count)]
print(files)
```

```python
# Create the frames
frames = [Image.open(i) for i in files]

# Save into a GIF file
frames[0].save('3d_vis.gif', format='GIF', append_images=frames[1:], save_all=True, duration=60, loop=0)
```
